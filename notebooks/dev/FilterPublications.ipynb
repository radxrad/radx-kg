```python
pis = pd.read_csv("../data/principal_investigators.csv")
pis = pis[["profile_id", "first_name", "middle_name", "last_name", "full_name", "project_serial_num", "last_name_initials"]].copy()
pis.drop_duplicates(inplace=True)
```

```python
dfm = subset.merge(pis, on="project_serial_num")
```

```python
dfm.head()
```

```python
from Levenshtein import jaro_winkler
def pi_match_score(authors, pi):
    """Return the best name match score (0-1) for a PI with a list of authors"""
    score = 0
    for author in authors:
         score = max(score, jaro_winkler(pi.strip(), author.strip()))

    return score
```

```python
dfm["pi_author"] = dfm[["authors", "full_name", "last_name_initials"]].apply(lambda x: max(pi_match_score(x["authors"], x["full_name"]), pi_match_score(x["authors"], x["last_name_initials"])), axis=1)
```

```python
#dfm.drop_duplicates(inplace=True)
```

```python
match_threshold = 0.85
dfm.query("pi_author >= 0.85", inplace=True)
dfm.sort_values("pi_author", ascending=False)
dfm.drop_duplicates("pm_id", inplace=True)
```

```python
df["pi_authored"] = df["pm_id"].isin(pi_authored_papers)
```

```python
counts_orig = pd.DataFrame(df["project_serial_num"].value_counts())
counts_orig.reset_index(names="project_serial_num", inplace=True)
counts_orig.rename(columns={"count": "total_publications"}, inplace=True)
```

```python
df_filtered = df[df["relevant"] | df["pi_authored"]]
```

```python
counts_filtered = pd.DataFrame(df_filtered["project_serial_num"].value_counts())
counts_filtered.reset_index(names="project_serial_num", inplace=True)
counts_filtered.rename(columns={"count": "filtered_publications"}, inplace=True)
```

```python
counts = counts_orig.merge(counts_filtered, on="project_serial_num", how="outer")
counts.fillna(0, inplace=True)
counts["filtered_publications"] = counts["filtered_publications"].astype(int)
counts.columns
```
