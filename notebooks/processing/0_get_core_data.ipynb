```python
projects = pd.read_csv(
        "https://docs.google.com/spreadsheets/d/1ZYUmUJEED3X6Mdo_sTMzOtIRW8IjsqWkAeLBKWdb31k/export?format=csv",
        keep_default_na=False,
       )
# duplicate, this phs number does not exist
projects.query("Accession != 'phs002574.v1.p1'", inplace=True)
```

```python
projects = projects[["Accession", "sub_project"]].copy()
```

## Prepare Grant Data

```python
grants = df.merge(projects, left_on="phs", right_on="Accession", how="right")
```

```python
grants
```

```python
grants.rename(columns={"DCC": "researchInitiative", "NIH RePORTER Grant Number ": "projectNum", "sub_project": "subProject"}, inplace=True)
```

```python
grants = grants[["researchInitiative", "projectNum", "subProject"]].copy()
```

```python
grants["projectNum"] = grants["projectNum"].str.split(";")
```

```python
grants = grants.explode("projectNum")
grants["projectNum"] = grants["projectNum"].str.strip()
```

```python
# Make corrections
```

```python
# remove invalid project number (Douglas Bell)
grants.query("projectNum != 'Z01-ES100475-07'", inplace=True)
```
